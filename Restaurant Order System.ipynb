The program takes input of item and quantity and does all calculation to provide the order receipt. In Restaurant class, login is validated with correct username and password. After that the menu is printed. We then take inputs and process order as per item and quantity given and print the sales receipt.

```python
class Restaurant:

    def __init__(self, name, menu):
        self.name = name
        self.menu = menu
        self.items_ordered = {}
        # managers: id and password pairs
        # i.e., id: lion and password: king
        self.managers = {"james":"bond", "lion":"king", "peter":"pan"}
        self.current_manager = ""

    def is_valid_login(self):        
        """ is valid login:
            take input from user, id and pw
            check id exists in self.managers
            check pw matches key value in self.managers
            max 3 tries
            return True or False
            if True, set self.current_manager if successful login
        """
        # write code here
        count = 0
        while True:
            userid  = (input("<<<"))
            pwd = (input("<<<"))
            if userid in self.managers.keys() and pwd in self.managers.values():                
                self.current_manager = userid
                break
            else:
                count += 1
                if count == 3:
                    return False
               
        return True
        
         
    def print_menu(self):
        """ print menu after successful login:
        
            --SUISP Ordering System--
            Wlecome to, SHAKE SHOCK BURGER!
            Current manager, Lion.

            ------------
            The menu
            ------------
            [1] Shock Burger $4.85
            [2] French Fries $2.85
            [3] Milk Shake $4.5
        """
        # write code here
        print ("--SUISP Ordering System--")
        print("Welcome to, SHAKE SHOCK BURGER!")
        print("Current manager is, " + self.current_manager + "\n")
        print ("------------")        
        print ("The Menu")
        print ("------------")
        for key, val in enumerate(self.menu):
            print ("[{}] {} ${}".format(key+1, val[0], val[1]))
            
        print ("User Id:", self.current_manager)        
        print ("Password:", self.managers[self.current_manager])
        print("Success!")

    def take_order(self, item, quantity):
        """ take order:
            item number and quantity
            update attribute self.items_ordered, i.e., {1:2, 3:1}
        """
        # write code here
        self.items_ordered[item] = quantity
            
        return self.items_ordered
        #print (self.items_ordered)
    def process_order(self):
        """ process order:
            use attribute, items_ordered
            calculate sub total price on all items
            return sub total
        """
        # write code here
        
        tax = 0.1    # 10% tax
        total = 0
        subtotal = 0
        
        for key, value in self.items_ordered.items():
            price = self.menu[key][1]
            
            subtotal += price * value
            
            calc_tax = subtotal * tax
            total = subtotal + calc_tax
            
        print("SubTotal: ${:.2f}".format(subtotal))
        print("Tax: ${:.2f}".format(calc_tax))
        print("Total: ${:.2f}".format(total))
              
        return total
        
        

    def print_sales_receipt(self):
        """ print sales receipt on this order:
            ++++++++++++++++++++++++
            The current order summary:
            ++++++++++++++++++++++++
            Shock Burger $4.85
                @ 3: item total $14.55
            French Fries $2.85
                @ 1: item total $2.85
            Milk Shake $4.50
                @ 5: item total $22.50
            ---------------------
            Sub Total: $39.90
            Tax: $3.99
            Total: $43.89
        """
#         # Sample code, printing all menu items
#         # How to traverse a dictionary with enumerate()
#         for k, v in enumerate(self.menu):
#             print ("{} ${:.2f}".format(v[0], v[1]))

        # write code here
        for k, v in self.items_ordered.items():    
            print("{} ${:.2f}".format(self.menu[k][0], self.menu[k][1]))
            print("   @ {}: item total ${:.2f}".format(self.items_ordered[k], self.items_ordered[k] * self.menu[k][1]))

        print ("---------------------")
        total = self.process_order()
        
             

def main():
    menu = ["Shock Burger", float(4.85), 50], ["French Fries", float(2.85), 200], ["Milk Shake", float(4.50), 20]
    rt = Restaurant('Shake Shock Burger', menu)

    if rt.is_valid_login():
        rt.print_menu()
        more_items = "yes"
        while more_items == "yes":
            choice, quantity = int(input("\nWhat would you like?\n")), int(input("How many would you like?\n"))
            if 0 <= (choice-1) < len(rt.menu):
                rt.take_order(choice-1, quantity)
            more_items = (input("Do you want to order more items?")).lower()
        rt.print_sales_receipt()
    else:
        print("System closed due to incorrect tries!")
        

if __name__ == '__main__':
    main()
```
